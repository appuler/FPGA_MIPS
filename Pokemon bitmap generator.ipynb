```python
png_files = []
file_names = []
for root, dirs, files in os.walk(current_dir):
    for file in files:
        if file.lower().endswith('.png'):
            full_path = os.path.join(root, file)
            file_names.append(file)
            png_files.append(full_path)
```

```python
newsize = (128, 128)
crop_img_dict = dict()
k = 0
for file in png_files:
    sub_images = []
    img = Image.open(file)
    img.resize(newsize)
    for i in range(8):
        for j in range(8):
            crop_box = (i*16, j*16, (i+1)*16, (j+1)*16)
            crop_img = img.crop(crop_box)
            sub_images.append(crop_img)
    
    crop_img_dict[file_names[k]] = sub_images
    k+=1
```

```python
for key in crop_img_dict.keys():
    new_key = key.replace(".png", "")
    os.mkdir(new_key)
    os.chdir(new_key)
    k = 0
    for i in crop_img_dict[key]:
        i.save(f"{k}.png")
        k += 1
    os.chdir("..")
```

```python
def compare_pngs(img_path1, img_path2, save_diff_path=None):
    # Open both images
    img1 = Image.open(img_path1).convert("RGB")
    img2 = Image.open(img_path2).convert("RGB")

    # Ensure images are the same size
    if img1.size != img2.size:
        raise ValueError("Images must be the same size to compare.")

    # Create a diff image
    diff = ImageChops.difference(img1, img2)

    # Count non-zero pixels (differences)
    diff_array = np.array(diff)
    num_diff_pixels = np.count_nonzero(diff_array)

    if save_diff_path:
        # Highlight differences in red
        diff_highlight = Image.new("RGB", img1.size)
        red = (255, 0, 0)
        mask = np.any(diff_array != 0, axis=-1)
        diff_highlight_array = np.array(diff_highlight)
        diff_highlight_array[mask] = red
        diff_highlight = Image.fromarray(diff_highlight_array)
        diff_highlight.save(save_diff_path)

    return num_diff_pixels
```

```python
def rgb_to_4bit(value):
    return round(value / 255 * 15)

def convert_png_to_bitmap_lines(image_path):
    img = Image.open(image_path).convert("RGB")
    width, height = img.size
    pixels = img.load()

    lines = []
    for y in range(height):
        row = []
        for x in range(width):
            r, g, b = pixels[x, y]
            r4, g4, b4 = rgb_to_4bit(r), rgb_to_4bit(g), rgb_to_4bit(b)
            row.append(f"{r4:X}{g4:X}{b4:X}")
        lines.append("\n".join(row))
    return lines


def batch_convert_to_single_file(folder, output_file="all_bitmaps.txt"):
    png_files = [f for f in os.listdir(folder) if f.lower().endswith(".png")]

    with open(output_file, "w") as out:
        for filename in png_files:
            img_path = os.path.join(folder, filename)
            lines = convert_png_to_bitmap_lines(img_path)

            out.write("\n".join(lines)) # Spacer between images
            out.write("\n")

    print(f"All bitmap data saved to: {output_file}")

# Example usage



os.chdir("C:\\Users\\Adi\\Desktop\\COMP\\Comp 541\\comp541Final\\pokemon")
for i in range(len(file_names)):
    folder = file_names[i].replace(".png", "")
    batch_convert_to_single_file(folder, f"{folder}/bitmap.txt")
```

```python
import os
import re

folder_path = os.getcwd()  # <- change this to your folder

pattern = re.compile(r'row-(\d+)-column-(\d+)\.png')

for filename in os.listdir(folder_path):
    match = pattern.match(filename)
    if match:
        y = int(match.group(1)) - 1  # row number (0-based)
        x = int(match.group(2)) - 1  # column number (0-based)
        i = 8 * y + x
        old_path = os.path.join(folder_path, filename)
        new_filename = f"{i}.png"
        new_path = os.path.join(folder_path, new_filename)
        os.rename(old_path, new_path)
        print(f"Renamed: {filename} -> {new_filename}")
```
